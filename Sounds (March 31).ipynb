```python
# Link for tutorial: https://publish.illinois.edu/augmentedlistening/tutorials/music-processing/tutorial-1-introduction-to-audio-processing-in-python/
# Here I'm importing stuff I need for this tutorial
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read, write
from IPython.display import Audio
from numpy.fft import fft, ifft
# 'Magic' funtion in python. Line-oriented. Sets the backend of matplotlib to the backend of inline
%matplotlib inline
```

```python
# Reading a .wav file and finding out its frequency
Fs, data = read('metallic-drums.wav')
# Here we are only using only one channel to keepp it simple. We are using channel 0
data = data[:,0]
print ('Sampling frequency is ', Fs)
```

```python
# This essentially 'displays' the audio
Audio(data, rate=Fs/8)
```

```python
# Plotting the frequency here!

plt.figure()
plt.plot(data)
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.title('Waveform of Test Audio')
plt.show()
```

```python
# We are writing the audio clip into another one.
write('output.wav', Fs, data)
```
